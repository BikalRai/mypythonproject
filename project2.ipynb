```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv

url = 'http://www.nepalstock.com/todaysprice'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'lxml')
soup
```

```python
tr = soup.table.find_all('tr')
tr
```

```python
data = []

for i in tr:
    a = i.get_text()
    data.append(a)
    
print(data)
```

```python
data1 = []

for i in data:
    b = i.split('\n')
    data1.append(b)
    
data1
```

```python
del data1[0]
data1
```

```python
del data1[-4:]
data1
```

```python
for a in data1:
    for b in a:
        if '' in b:
            b = b.replace(",", "")
            f = open('project2.csv', 'w', encoding = 'utf-8')
            write = csv.writer(f)
            for i in data1:
                write.writerow(i)
    
f.close()
            
```

```python
df = pd.read_csv('project2.csv')
df
```

```python
import plotly.express as px

fig = px.pie(df, values='No. Of Transaction', names='Traded Companies')
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')
fig.show()
```

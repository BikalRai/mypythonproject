```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import csv

url = "https://www.worldometers.info/coronavirus/"
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'lxml')
soup
```

```python
data = soup.table.find_all('tr')
data
```

```python
data1 = []

for i in data:
    x = i.get_text()
    data1.append(x)
    
print(data1)
```

```python
data2 = []

for i in data1:
    y = i.split('\n')
    data2.append(y)
    
data2
```

```python
del data2[1:9]
data2
```

```python
del data2[-8:]
data2
```

```python
f = open('Covid data.csv', 'w')
write = csv.writer(f)
for i in data2:
    write.writerow(i)

f.close()
```

```python
df = pd.read_csv('Covid data.csv', index_col = "#")
df
```

```python
import plotly.express as px
fig = px.pie(df, values='TotalDeaths', names='Country,Other',
             title='COVID-19 CORONAVIRUS PANDEMIC',
             hover_data=['TotalCases'], labels={'TotalDeaths'})
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
